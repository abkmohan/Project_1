```python
# Problem statement 8  - Make country as the first column of the dataframe
columnlist = df.columns.tolist()
countryIndex = columnlist.index("Country")
columnlist.pop(countryIndex)
columnlist.insert(0, 'Country')
df = df[columnlist]  #displaying with "Country" as the first column
df

#Additional mentor given exercises
#Distribution of categories like (Country, Published, Year) (Distribution of each category)
CountryD = df.groupby(['Country'])['Country'].count().reset_index(name='Countrydistribution')
CountryD['Countrydistribution'] = CountryD['Countrydistribution'] / CountryD['Countrydistribution'].sum()
CountryD

YearD = df.groupby(['Year'])['Year'].count().reset_index(name='Yeardistribution')
YearD['Yeardistribution'] = YearD['Yeardistribution'] / YearD['Yeardistribution'].sum()
YearD

PublishedD = df.groupby(['PUBLISH STATES'])['PUBLISH STATES'].count().reset_index(name='Yeardistribution')
PublishedD['Yeardistribution'] = PublishedD['Yeardistribution'] / PublishedD['Yeardistribution'].sum()
PublishedD

#Distribution of two categories - (Year Vs Country),(Year Vs Status),(Year Vs Numeric)
CountryYearD = df.groupby(['Country','Year'])['Country'].count().reset_index(name='Countrydistribution')
CountryYearD['Countrydistribution'] = CountryYearD['Countrydistribution'] / CountryYearD['Countrydistribution'].sum()
CountryYearD

StatusYearD = df.groupby(['Year', 'PUBLISH STATES'])['Year'].count().reset_index(name='Yeardistribution')
StatusYearD['Yeardistribution'] = StatusYearD['Yeardistribution'] / StatusYearD['Yeardistribution'].sum()
StatusYearD

NumericYearD = df.groupby(['Year', 'Numeric'])['Year'].count().reset_index(name='Yeardistribution')
NumericYearD['Yeardistribution'] = NumericYearD['Yeardistribution'] / NumericYearD['Yeardistribution'].sum()
NumericYearD

```

```python

users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')

users.head()
sessions.head()
products.head()
transactions.head()
```

```python
# Problem Statement 12. - Join users to transactions, keeping all rows from transactions and only matching rows from users
#                    (left join)

from pandasql import sqldf

pysqldf = lambda statement: sqldf(statement, globals())
new_df = pysqldf("SELECT * from users as u LEFT OUTER JOIN transactions as t ON u.UserId = t.UserId;")
new_df
```

```python
# Problem Statement 13 -  Which transactions have a UserID not in users?
new_df = pysqldf("SELECT * from transactions as t where userid is null or userid not in (select userid from users as u);")
new_df
```

```python
# Problem Statement 14 -Join users to transactions, keeping only rows from transactions and users that
#                       match via UserID (inner join)
new_df = pysqldf("SELECT * from users as u INNER JOIN transactions as t on u.userid = t.userid;")
new_df
```

```python
# Problem statment 15. Join users to transactions, displaying all matching rows AND all non-matching rows
#                      (full outer join)
new_df = pysqldf("SELECT * from users,transactions;")
new_df
```

```python
# Problem statement 16: Determine which sessions occurred on the same day each user registered
new_df = pysqldf("SELECT u.UserId, u.User, u.Gender, u.registered, u.cancelled, s.sessionid, s.sessiondate from sessions as s inner join users as u group by u.registered;")
new_df
```

```python
# Problem statement 17: Build a dataset with every possible (UserID, ProductID) pair (cross join)
new_df = pysqldf("SELECT UserId, productid from users,products")
new_df
```

```python
# Problem statement 18: Determine how much quantity of each product was purchased by each user
new_df = pysqldf("""
            SELECT u.userid, p.productid, sum(t.quantity) as 'totalquantity' 
            FROM users as u JOIN products as p
                LEFT OUTER JOIN transactions as t on u.userid = t.userid                
            GROUP BY u.userid, p.productid 
            """)
new_df.fillna(0)
```

```python
# Problem statement 19: For each user, get each possible pair of pair transactions (TransactionID1,TransacationID2)
tran_df = pysqldf("""
            SELECT t1.TransactionID as 'TransactionID_x'
                    ,t1.TransactionDate as 'TransactionDate_x'
                    ,t1.UserID as 'UserID_x'
                    ,t1.ProductID as 'ProductID_x'
                    ,t1.Quantity as 'Quantity_x'
                    ,t2.TransactionID as 'TransactionID_y'
                    ,t2.TransactionDate as 'TransactionDate_y'
                    ,t2.UserID as 'UserID_y'
                    ,t2.ProductID as 'ProductID_y'
                    ,t2.Quantity as 'Quantity_y'
            FROM transactions as t1 JOIN transactions as t2 on t1.userid = t2.userid
            """)
tran_df
```

```python
# Problem Statement 20: Join each user to his/her first occuring transaction in the transactions table
data = pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
data
```
